```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('E://Gaurav/ML_Dataset/Multiple_Linear_Regression/50_.csv')
dataset
```

```python
x = dataset.iloc[:,:-1].values
y =dataset.iloc[:,4].values  # dependednt variable at 4th column
```

```python
x # Matrix of multiple independent Variables
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_x = LabelEncoder()
labelencoder_x.fit_transform(x[:,3]) # first column encoded  # the outputis the encoded value
```

```python
# applyiong to the original array

x[:,3] = labelencoder_x.fit_transform(x[:,3]) 
```

```python
x
```

```python
# now, after encoding, we need to convert it into dummy variable using OneHotEncoder:

from sklearn.preprocessing import OneHotEncoder
```

```python
onehotencoder = OneHotEncoder(categorical_features = [3] ) 
x = onehotencoder.fit_transform(x).toarray()   
```

```python
x
```

```python
# Dummy Variable Trap:

x = x[:,1:] # since 2 variables here, so we can have just 1 dummy variable
```

```python
x
```

```python
# dividing the data into training and test data:

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```
